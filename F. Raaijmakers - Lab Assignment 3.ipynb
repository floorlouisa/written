```python
def compute_centroid(x,mu):
    x = reshape(x,(64,1))
    mu = mu - x
    n = np.sqrt(np.sum(mu*mu, axis=0))
    return argmin(n)
```

```python
def update_centroid(centroid, mu, X_c):
    mean = np.mean(X_c, axis=0)
    mu[:,centroid] = mean
    return mu
```

```python
def cost_function(x,mu):
    x = reshape(x,(64,1))
    mu = mu - x
    n = np.sqrt(np.sum(mu*mu, axis=0))
    return n[argmin(n)]
```

```python
#-----Initialization------#
initialisation_iterations = 10
iterations = 100

k = 10 #10 digits

x = np.reshape(digits.images[:1500],(1500,64))
x_test = np.reshape(digits.images[1500:],(297,64))

#----K-Means Algorithm----#
for h in range(initialisation_iterations): #different initialisations of mu
    mu = np.random.uniform(0,16,(64,k))
    c = np.random.randint(0,k,(1500))
    cost = [] #initilize cost array to save cost values to
    for j in range(iterations): #amount of iterations of the k means algorithm
        i_cost = 0 #initialize current cost
        for i in range(k):
             mu = update_centroid(i,mu,x[c==i]) #updates the centroid of all the clusters
        for i in range(x.shape[0]):               
            c[i] = compute_centroid(x[i,:],mu)  #assigns a new cluster to each datapoint
            i_cost  += cost_function(x[i,:],mu) #updates current cost
        i_cost = i_cost/x.shape[0]
        cost.append(i_cost) 
    
    #---Plotting---#
    plt.plot(range(iterations),cost)
    plt.xlabel('Iteration')
    plt.ylabel('Cost')
    print "Iteration", h+1
    plt.show()

visualize(c)
visualize_centroids(mu)
        
```

## Analysis
As we increase the number of iterations we observe that the clusters become more and more accurate, meaning less digits are 
categorized to the wrong cluster. However, we want to keep a balance between number of iterations and the marginal effect that an 
extra number of iterations has on the purity of the cluster. I have observed that for this data set, a 100 iterations is a good
trade-off between accuracy and the time it taken. Furthermore, from the cost plots we can observe that the more initializations, 
the quicker the cost drops. Hence it is neccesary to have many initializations.
